```python
def evaluar_año(fecha_str):
    año = int(fecha_str[-4:])
    return año < 2024

df["próximoavencer"] = df["fechavencimiento"].apply(evaluar_año)
```

```python
df.drop_duplicates(inplace=True)
```

```python
eliminar_columnas=["atc","expediente"]
df = df.drop(eliminar_columnas, axis=1)
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
from unidecode import unidecode

def eliminar_acentos(columna):
    return columna.apply(lambda x: unidecode(x) if isinstance(x, str) else x)

df = df.apply(eliminar_acentos)
```

```python
import matplotlib.pyplot as plt
```

```python
df_temp = df.copy()
df_temp["año de vencimiento"] = df_temp["fechavencimiento"].astype(str).str[-4:].astype(int)
df_temp["es.for.far en numerico"] = df_temp["estadodeformafarmaceutica"].map({"sólido": 0, "semisólido": 1, "líquido": 2, "gas": 3, "sistema heterodisperso": 4, "otras/especiales": 5})
correlation = df_temp["es.for.far en numerico"].corr(df_temp["año de vencimiento"])
print(f"Coeficiente de correlación:{correlation}")
```

```python
plt.scatter(df_temp["año de vencimiento"], df_temp["es.for.far en numerico"])
plt.xlabel("año de vencimiento")
plt.ylabel("es.for.far en numerico")
plt.title("Correlación entre año de vencimiento y estado de forma farmacéutica en valores numericos")
plt.show()
```

```python
correlation=df_temp["es.for.far en numerico"].corr(df_temp["próximoavencer"])
print(f"Coeficiente de correlación:{correlation}")
```
